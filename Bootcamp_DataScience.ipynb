### Desafio 05: Buscar os casos de dengue no Brasil (época de maior número de casos e regiões mais atingidas) e se os picos de alguns estados em fevereiro e verão de modo geral, pode ser reflexos dos casos de dengue

Obtive na site Tabnet os dados de relativos a dengue por região dos anos de 2008a 20013.
http://tabnet.datasus.gov.br/cgi/tabcgi.exe?sinannet/cnv/denguebr.def
Essas informações serão comparadas com os custos da saúde por região no mesmo período.

```
#Selecionando dados
list_coluns_regiao= usaveis.columns[:71].tolist()
list_coluns_regiao.append("Regiao")
dados_regiao_2008_2013 = usaveis[list_coluns_regiao]
#Obtendo Total por Região
totais_por_regiao = dados_regiao_2008_2013.groupby(by=["Regiao"]).sum()

#Plot do Gráfico
axis = totais_por_regiao.T.plot(figsize = (10,8), x_compat = True, colormap = 'Dark2', fontsize = 12)
axis.yaxis.set_major_formatter(ticker.StrMethodFormatter("R$ {x:,.2f}"))
plt.xticks(rotation=45)
plt.legend(loc = 'right', bbox_to_anchor=(1.15, 0.8), fontsize = 14)
```

```
#Tratando os dados de Dengue
#Essa dataSet dos dados da Dengue tem o total do período de 2008 - 2013 por Mês e Região
dados_dengue = pd.read_csv("/content/Dados_degue_2008-2013_regiao.csv", skiprows=3, 
                            sep=";", skipfooter=16, thousands=".", decimal=",")
            
dados_dengue = dados_dengue.drop("Total", axis = 1 )
dados_dengue = dados_dengue.drop([13])

#ajustar nomes região
dados_dengue = dados_dengue.T.rename({"M�s 1� Sintoma(s)" : "Mês Sintomas", "1 Regi�o Norte": "Norte",
                       "2 Regi�o Nordeste":"Nordeste", "3 Regi�o Sudeste":"Sudeste",
                       "4 Regi�o Sul":"Sul", "5 Regi�o Centro-Oeste":"Centro-Oeste"}, axis = 'index')
#Retirar linhas sem valores relevantes
dados_dengue = dados_dengue.T.drop (0)

#Plot do Gráfico
dados_dengue = dados_dengue.set_index('Mês Sintomas')
dados_dengue = dados_dengue.astype(float)
axis = dados_dengue.plot(figsize = (10,8), x_compat = True, colormap = 'Dark2', fontsize = 12)
axis.set_xticks(np.arange(0,12))
axis.set_xticklabels(dados_dengue.T.columns)
axis.xaxis.grid(True)

plt.xticks(rotation=45)
plt.xlabel("Meses")
plt.ylabel("Montante acumulado anos 2008-2013")
plt.title ("Total Montante 2008-2013 por Região Brasil", fontsize = 14)
plt.legend(loc = 'right', bbox_to_anchor=(1.30, 0.8), fontsize = 14)
```

```
ind_regiao =totais_por_regiao.index
ind_meses =dados_dengue.index
#Criação DataFrame dos gastos totais acumulado de 2008-2013
gastos_Totais = pd.DataFrame(index =ind_regiao, columns=ind_meses )
#gastos_Totais

gastos_Totais["Janeiro"] = totais_por_regiao[['2008/Jan','2009/Jan','2010/Jan','2011/Jan','2012/Jan','2013/Jan']].sum(axis = 1)
gastos_Totais["Fevereiro"] = totais_por_regiao[['2008/Fev','2009/Fev','2010/Fev','2011/Fev','2012/Fev','2013/Fev']].sum(axis = 1)
gastos_Totais["Marco"] = totais_por_regiao[['2008/Mar','2009/Mar','2010/Mar','2011/Mar','2012/Mar','2013/Mar']].sum(axis = 1)
gastos_Totais["Abril"] = totais_por_regiao[['2008/Abr','2009/Abr','2010/Abr','2011/Abr','2012/Abr','2013/Abr']].sum(axis = 1)
gastos_Totais["Maio"] = totais_por_regiao[['2008/Mai','2009/Mai','2010/Mai','2011/Mai','2012/Mai','2013/Mai']].sum(axis = 1)
gastos_Totais["Junho"] = totais_por_regiao[['2008/Jun','2009/Jun','2010/Jun','2011/Jun','2012/Jun','2013/Jun']].sum(axis = 1)
gastos_Totais["Julho"] = totais_por_regiao[['2008/Jul','2009/Jul','2010/Jul','2011/Jul','2012/Jul','2013/Jul']].sum(axis = 1)
gastos_Totais["Agosto"] = totais_por_regiao[['2008/Ago','2009/Ago','2010/Ago','2011/Ago','2012/Ago','2013/Ago']].sum(axis = 1)
gastos_Totais["Setembro"] = totais_por_regiao[['2008/Set','2008/Set','2010/Set','2011/Set','2012/Set','2013/Set']].sum(axis = 1)
gastos_Totais["Outubro"] = totais_por_regiao[['2008/Out','2009/Out','2010/Out','2011/Out','2012/Out','2013/Out']].sum(axis = 1)
gastos_Totais["Novembro"] = totais_por_regiao[['2008/Nov','2009/Nov','2010/Nov','2011/Nov','2012/Nov','2013/Nov']].sum(axis = 1)
gastos_Totais["Dezembro"] = totais_por_regiao[['2008/Dez','2009/Dez','2010/Dez','2011/Dez','2012/Dez','2013/Dez']].sum(axis = 1)

gastos_Totais
```

```
x = dados_dengue.index
y1 = gastos_Totais.loc['SE']
y2 = dados_dengue.T.loc['Sudeste']

#Criar a área de plotagem e os dois eixos y
fig, ax1 = plt.subplots(figsize=(10,8))
plt.xticks( rotation = 90)
ax2 = ax1.twinx()

curva1, = ax1.plot(x, y1, label='Gastos da região SE', color='k', marker=".", markersize=15)
curva2, = ax2.plot(x, y2, label='Casos de dengue em SE', color='b',  marker=".", markersize=15)

#Criar uma lista para acessar os parametros das curvas
curvas = [curva1, curva2]

#Criar a legenda do gráfico
ax1.legend(curvas, [curve.get_label() for curve in curvas], loc = 'best', bbox_to_anchor=(1.30, 0.8), fontsize = 14)

#Criar os labels de cada eixo
ax1.set_xlabel('Meses', {'size':12})
ax1.set_ylabel('Valores em milhões de reais', {'size':12}, color=curva1.get_color())
ax2.set_ylabel('Casos de dengue', {'size':12}, color=curva2.get_color())

#Alterar a cor dos eixos y
ax1.tick_params(axis='y', colors=curva1.get_color())
ax2.tick_params(axis='y', colors=curva2.get_color())

#Setar os valores dos eixos y
ax1.yaxis.set_major_formatter(ticker.StrMethodFormatter('R$ {x:,.0f}'))

plt.title('Comparativo anual de gastos\ncom saúde X Casos de dengue\nRegão SE', {'size':14})
plt.show()
```

##Considerações sobre Desafio 5
Foram obtidos os casos acumulados de dengue de 2008 a 2013 separados pelos meses do ano, de cada região do Brasil. Também foram extraidos informações acumulados dos gastas da saúde no Braisil por região divididas pelos meses do ano.

No gráfigo plotado, que se delimita a região Sudeste pode-se verficar uma tendência de crescimento na curva de cor azul, referente ao acumulado de casos de degue, no período de Fevereio e Março. Esta tendência também é observado na curva de cor preta, referente ao montante de gastos na Saúde no Sudeste do Brasil, no mesmo período de Feveiro a Março.
Nos posterioresa Março, a tendência da curva de casos de dengue é decair e estabilizar após o mês de Junho, mantendo-se até Dezembro. Na curva de gastos, a diferença entre os gastos dos meses é menor que a diferença entre os meses de Fevereiro e Março.

Portanto, pode-se supor que há uma relação proporional entre o aumento dos casos de dengue e os gastos da Saúde no Sudeste do Brasil nos meses de Fevereiro e Março.

O método de acumular os casos de dengue e os montantes de gastos nos meses entre os anos de 2008 e 2013 podem ocultar informações não relacionadas a correlação pressuposta, entretanto são indícios que pode existir um relação entre os gastos da saúde pública e os casos de degue. Para aumetar a acurácia dos estudo deveriam ser obtidos mais informações sobre outras doenças no mesmo período.
