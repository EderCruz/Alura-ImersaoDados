## Desafio 01: Escolher um título mais descritivo, que passe a mensagem adequada ao gráfico de barras.

```
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

axis = dados.plot(x="Unidade da Federação",y="2008/Ago", kind = "bar", figsize = (10,8))
axis.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))

plt.title ("Montante gastos da saúde Ago/2008 por Unidade Federativa")
plt.show()
```

### Desafio 02: Faça a mesma análise realizada em aula, porém para o mês mais recente.

```
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

axis = dados.plot(x="Unidade da Federação",y="2020/Jul", kind = "bar", figsize = (10,8))
axis.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))

plt.title ("Montante gastos da saúde Jul/2020 por Unidade Federativa")
plt.show()
```

### Desafio 01: Reposicionar a legenda do gráfico em uma posição mais adequada

```
usaveis.T.plot(figsize=(12,8))
plt.legend(loc = 'right', bbox_to_anchor=(1.30, 0.5))
plt.show
```

###Desafio 02: Plotar o gráfico de linha com apenas 5 estados de sua preferência

```
estados_minha_preferencia_NORTE_BR = usaveis.T[["15 Pará", "12 Acre", "13 Amazonas", "14 Roraima", "17 Tocantins"]]
axis = estados_minha_preferencia_NORTE_BR.plot(figsize = (10,8))
axis.yaxis.set_major_formatter(ticker.StrMethodFormatter("R$ {x:,.2f}"))
plt.legend(loc = 'right', bbox_to_anchor=(1.22, 0.5))
plt.show
```
