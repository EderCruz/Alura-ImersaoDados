Definição de funções que serão utilizadas para tratamento dos dados  neste projeto.

```
def ajustarTabelaTabNte (tabAlvo):
  tabAlvo.set_index('Unidade da Federação', inplace = True);
  tabAlvo.drop('Total', axis = 0, inplace = True)
  tabAlvo.drop('Total', axis = 1, inplace = True)
  tabAlvo.index = tabAlvo.index.str[3:]
  
  return tabAlvo

def melt_tab(tabAlvo):
  tabAlvo = tabAlvo.T.reset_index().melt(id_vars = ["index"], value_vars=tabAlvo.T.columns )
  tabAlvo.columns = ["ano","uf","valor"]
  return tabAlvo

def catplot_melt_ano_uf_valor(dataPlotar, titulo, cor_paleta, tipo_plot):
  plt.figure(figsize=(8,6))
  sns.set(style="ticks")
  ax = sns.catplot(data=dataPlotar, x = 'ano', y = 'valor', kind=tipo_plot, col='uf', palette = cor_paleta)
  plt.subplots_adjust(top=0.85)
  ax.fig.suptitle(titulo, fontsize=18)
  for g in ax.axes.flat:
    g.yaxis.set_major_formatter(ticker.FuncFormatter(lambda y, p: f'{y:.0f}'))
    g.grid(color='r', linestyle='--', linewidth=1.5, axis='y')
  plt.show()
```

Importação do resultado das eleições 2018 por estado

```
from unicodedata import normalize

table_wiki_eleicoes_pres_2018 = pd.read_html('https://pt.wikipedia.org/wiki/Resultados_da_elei%C3%A7%C3%A3o_presidencial_no_Brasil_em_2018#:~:text=Nesta%20p%C3%A1gina%2C%20est%C3%A3o%20listados%20os,%2C87%25%20dos%20votos%20v%C3%A1lidos.'
, match = 'Estado')
df_eleicoes_2_turno_2018 = table_wiki_eleicoes_pres_2018[3]

df_eleicoes_2_turno_2018['Estados'] = df_eleicoes_2_turno_2018['Estado[3]']
df_eleicoes_2_turno_2018['%Bolsonaro'] = df_eleicoes_2_turno_2018['%.1']
df_eleicoes_2_turno_2018['%Haddad'] = df_eleicoes_2_turno_2018['%.2']
df_eleicoes_2_turno_2018 = df_eleicoes_2_turno_2018.drop(['Estado[3]', 'Abstenção', 'Votos brancos', 'Votos nulos', '%','%.1','%.2','%.3', '%.4'], axis= 1)
df_eleicoes_2_turno_2018 = df_eleicoes_2_turno_2018[['Estados', 'Bolsonaro', '%Bolsonaro','Haddad','%Haddad']]

df_eleicoes_2_turno_2018['Bolsonaro'] = df_eleicoes_2_turno_2018['Bolsonaro'].str.replace('\xa0',"").astype(int)
df_eleicoes_2_turno_2018['Haddad'] = df_eleicoes_2_turno_2018['Haddad'].str.replace('\xa0',"").astype(int)
df_eleicoes_2_turno_2018['%Bolsonaro'] = df_eleicoes_2_turno_2018['%Bolsonaro'].str.replace('%','')
df_eleicoes_2_turno_2018['%Bolsonaro'] = df_eleicoes_2_turno_2018['%Bolsonaro'].str.replace(',','.').astype(float)
df_eleicoes_2_turno_2018['%Haddad'] = df_eleicoes_2_turno_2018['%Haddad'].str.replace('%','')
df_eleicoes_2_turno_2018['%Haddad'] = df_eleicoes_2_turno_2018['%Haddad'].str.replace(',','.').astype(float)
df_eleicoes_2_turno_2018['Estados'] = df_eleicoes_2_turno_2018['Estados'].str.strip()

df_eleicoes_2_turno_2018.set_index('Estados', inplace=True)

df_eleicoes_2_turno_2018 = df_eleicoes_2_turno_2018.drop(['Total','Exterior'], axis= 0)
df_eleicoes_2_turno_2018.head()
```

##7.1 Fatores que influenciam na redução da Cobertura Vacinal no Brasil
Na revista **Consensus** do quarto trimestre 2017, publicada pela Fio Cruz, o artigo "A queda da imunização no Brasil"[13] discute alguns fatores que podem influenciar na cobertura vacinal no brasil:


*   **A importância da vacina:** felizmente, para a maioria da população significa não apenas proteção, mas sustentar uma condição de saúde coletiva 
*   **Movimentos antivacinas:** cresce o número de pessoas que se recusam a vacinar seus filhos, fomentando um movimento perigoso que pode trazer de volta doenças erradicadas.
*   **Jovens e Adultos:** Conscientizar à população jovem e adulta, para que entenda o calendário de vacinação não é só para as crianças, também é um desafio do programa de imunização.
*   **A importância da comunicação:** imprescindível aprimorar a comunicação com a população e com os profissionais de saúde,  nas Unidades
Básicas e com Agentes Comunitários de Saúde.

Também na revista eletrônica da fapespa **Pesquisa Fapespa** o artigo "As razões da queda da vacinação"[14] relata que pelo menos nove fatores contribuem para a redução na imunização infantil e o risco de doenças erradicadas surgirem.
